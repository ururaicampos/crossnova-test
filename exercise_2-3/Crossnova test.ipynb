## Exercise 2

The general equation of the plot is given by:
$$ f(t) = \text{A} t + \text{B} \sin (\omega_1 t + \varphi_1) + \text{C} \cos(\omega_2 t + \varphi_2) $$

Where $\text{A}$, $\text{B}$, $\text{C}$ are constant values and:

- $\omega = 2\pi \text{f}$, is the angular frequency in radians per second.
- $\text{f} = \frac{1}{T}$, is the ordinary frequency.
- $\varphi$, is the fase where in its cycle the oscillation is at t = 0.

```python
import numpy as np
import matplotlib.pyplot as plot


class Plotting:
    
    def __init__(self):
        self.time = np.arange(0, 14, 0.005)
        self.constants = [5, 0.5, 0.5]
        self.omegas = [-45, 50]
        self.phis = [0, -np.pi/2]
    
    def show_plot(self):
        dataset = self.time/self.constants[0] \
                + self.constants[1]*np.sin(self.omegas[0] * self.time + self.phis[0]) \
                + self.constants[2]*np.cos(self.omegas[1] * self.time + self.phis[1])
        plot.figure(figsize=(15,10))
        plot.plot(self.time, dataset)
        plot.xlabel('Time $(t)$')
        plot.ylabel('Amplitude')
        return plot.show()
        
Plotting().show_plot()
```

Mario brought **four boxes** with the same amount of $\text{n}$ toffees. 

$$ \text{y(n)} = 4 * \text{n} $$

For each box $\text{f}_i$ this mean that:

$$ \text{n} = \text{f}_1(x_1) = \text{f}_2(x_2) = \text{f}_3(x_3) = \text{f}_4(x_4) $$  

If Mario ate **one toffee** of each box and divided the rest with his $\text{x}$ friends. The total amount of each box is given by the following equation:

$$ 1 + \frac{n-1}{4} = 1 + \frac{n-1}{5} = 1 + \frac{n-1}{6} = 1 + \frac{n-1}{7}$$

$$ \frac{n-1}{4} = \frac{n-1}{5} = \frac{n-1}{6} = \frac{n-1}{7} $$

For the equation to be true, the only possible value for it is **n=1**. Meaning that Mario is a bad friend who ate the chocolate by himself and didn't gave it to his other friends. 

The total amount that Mario brought $\text{y(n)}$ is **4 toffees**.

$$ \text{y(1)} = 4 * 1 = 4$$
